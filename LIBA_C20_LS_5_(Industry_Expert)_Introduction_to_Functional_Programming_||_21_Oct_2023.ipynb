```python
# Chained Conditionals
x = 35
if x>35:
  print(x,"is greater than 35")
elif x<35:
  print(x,"is less than 35")
else:
  print(x,"is equal than 35")
```

```python
# Chained Conditionals
x = -35
if x>35:
  print(x,"is greater than 35")
elif x<35:
  print(x,"is less than 35")
else:
  print(x,"is equal than 35")
```
